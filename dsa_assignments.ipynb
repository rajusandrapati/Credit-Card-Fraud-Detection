```python
import pandas as pd
import numpy as np

# Sample DataFrame for demonstration
data = {
    'A': [1, np.nan, 3, 4, np.nan, 6],
    'B': [7, 8, np.nan, 10, 11, 12],
    'C': [15, 200, 25, 300, 35, 50]
}
df = pd.DataFrame(data)
print("Original DataFrame:\n", df)

# 1. Interpolate missing values in a DataFrame using a linear method
df_interpolated = df.interpolate(method='linear')
print("\nDataFrame after linear interpolation:\n", df_interpolated)

# 2. Use a mask created with NumPy to fill missing values in a DataFrame with a specific value
mask = np.isnan(df.to_numpy())  # Create a mask for missing values
df_filled = df.fillna(value=0)  # Fill missing values with 0
print("\nDataFrame after filling missing values with 0:\n", df_filled)

# 3. Identify and replace outliers in a DataFrame column with the median value using a NumPy mask
# Define outliers as values that are more than 1.5 times the IQR above the third quartile or below the first quartile
Q1 = df['C'].quantile(0.25)
Q3 = df['C'].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Create a mask for outliers
outliers_mask = (df['C'] < lower_bound) | (df['C'] > upper_bound)

# Replace outliers with the median value
median_value = df['C'].median()
df['C'] = np.where(outliers_mask, median_value, df['C'])
print("\nDataFrame after replacing outliers in column 'C' with median value:\n", df)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Sample DataFrame for demonstration
np.random.seed(0)
data = {
    'A': np.random.randn(100) * 10 + 50,  # Normally distributed data
    'B': np.random.rand(100) * 100,       # Uniformly distributed data
    'C': np.random.randint(1, 10, size=100)  # Random integers
}
df = pd.DataFrame(data)

# 26. Perform exploratory data analysis by calculating summary statistics
summary_stats = df.describe(include='all').T  # Transpose for better readability
summary_stats['median'] = df.median()
summary_stats['mode'] = df.mode().iloc[0]  # Mode can be multiple values, take the first one
summary_stats['standard deviation'] = df.std()
print("Summary Statistics:\n", summary_stats)

# 27. Visualize the distribution of a numerical column using a histogram and a box plot
plt.figure(figsize=(14, 6))

# Histogram
plt.subplot(1, 2, 1)
sns.histplot(df['A'], kde=True)
plt.title('Histogram of Column A')

# Box Plot
plt.subplot(1, 2, 2)
sns.boxplot(x=df['A'])
plt.title('Box Plot of Column A')

plt.show()

# Explanation:
# Outliers in the box plot are points outside the whiskers. They are extreme values that fall below Q1 - 1.5*IQR or above Q3 + 1.5*IQR.

# 28. Create a correlation matrix and visualize it using a heatmap
correlation_matrix = df.corr()
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Matrix Heatmap')
plt.show()

# Explanation:
# The heatmap shows the correlation between numerical columns. Values close to 1 or -1 indicate strong positive or negative correlations, respectively.

# 29. Create a scatter plot matrix of multiple columns and analyze relationships
sns.pairplot(df)
plt.suptitle('Scatter Plot Matrix', y=1.02)
plt.show()

# Explanation:
# The scatter plot matrix (or pair plot) shows relationships between pairs of numerical columns. It helps in understanding the correlation and distribution.

# 30. Perform feature engineering by creating new features and visualize their importance
df['A_squared'] = df['A'] ** 2  # Create a new feature
df['B_log'] = np.log(df['B'] + 1)  # Create a new feature, adding 1 to avoid log(0)

# Visualize feature importance
# For simplicity, we can use absolute values of correlations as a measure of importance
feature_importance = df.corr()['A'].drop('A')  # Correlation with column 'A' as an example
feature_importance.plot(kind='bar', title='Feature Importance')
plt.ylabel('Correlation with A')
plt.show()
```
